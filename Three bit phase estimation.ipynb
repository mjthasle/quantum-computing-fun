```python
from qiskit import *
import numpy as np
%matplotlib inline

circuit = QuantumCircuit(4, 3)
circuit.h(0)
circuit.h(1)
circuit.h(2)
circuit.rx(np.pi/2,3)
circuit.cry(np.pi,0,3)
circuit.cry(2*np.pi,1,3)
circuit.cry(4*np.pi,2,3)
circuit.h(0)
circuit.h(1)
circuit.h(2)
circuit.crx(-np.pi/2,2,1)
circuit.crx(-np.pi/4,2,0)
circuit.crx(-np.pi/2,1,0)
circuit.measure([0,1,2],[2,1,0])
circuit.draw('mpl')
```

```python
provider = IBMQ.load_account()
simulator_backend = Aer.get_backend('qasm_simulator')
# Use 16 qubit Melbourne as real backend
real_backend = provider.get_backend('ibmq_athens')
```

```python
job = execute(circuit, simulator_backend, shots=1000)
counts = job.result().get_counts(circuit)
from qiskit.visualization import plot_histogram
plot_histogram(counts)
```

```python
real_backend = provider.get_backend('ibmqx2')
job = execute(circuit, real_backend, shots=1000)
counts  = job.result().get_counts(circuit)
from qiskit.visualization import plot_histogram
plot_histogram(counts)
```
